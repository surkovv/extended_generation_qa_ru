```python
def get_tree(conllu_output):
    tree = Conllu(filehandle=StringIO(conllu_output)).read_tree()
    return tree

def get_trees(conllu_output):
    tree = Conllu(filehandle=StringIO('\n\n'.join(conllu_output))).read_trees()
    return tree

def write_tree(conllu_output):
    tree = get_tree(conllu_output)
    writer = TextModeTrees(attributes="ord,form,lemma,upos,deprel,parent,feats")
    writer.process_tree(tree)
    
```

```python
from collections import namedtuple
```

```python
Instance = namedtuple('Instance', ['question', 'question_tree', 'answer', 'answer_tree'])

def getInstances(records, chunk_size=50):
    instances = []
    for chunk_start in tqdm(range(0, len(records), chunk_size)):
        chunk_end = min(chunk_start + chunk_size, len(records))
        chunk = records[chunk_start:chunk_end]
        
        questions = [record.question for record in chunk]
        question_trees = get_trees(parser(questions))
        
        answers = [record.answer for record in chunk]
        answer_trees = get_trees(parser(answers))
        
        instances.extend([Instance(*x) for x in zip(questions,
                                                   question_trees,
                                                   answers,
                                                   answer_trees)])
    return instances
```

```python
instances = getInstances(records)
```

```python
def find_sub(question_tree, find_list):
    for dec in question_tree.descendants(add_self=True):
        if dec.form.lower() in find_list:
            return True
    return False

def who_what_nom(question_tree):
    root = question_tree.children[0]
    if root.form.lower() in ['', '']:
        return True
    return False

def who_what(question_tree):
    root = question_tree.children[0]
    if len(root.children) > 0:
        clause = root.children[0]
        for word in clause.descendants(add_self=True):
            if word.lemma in ['', '', '']:
                return True
    return False

def which_what(question_tree):
    root = question_tree.children[0]
    if len(root.children) > 0:
        clause = root.children[0]
        for word in clause.descendants(add_self=True):
            if word.lemma in ['', '', '', '', '']:
                return True
    return False

def kakovo(question_tree):
    root = question_tree.children[0]
    if root.lemma in ['']:
        return True
    return False

def time(question_tree):
    root = question_tree.children[0]
    if len(root.children) > 0:
        clause = root.children[0]
        for word in clause.descendants(add_self=True):
            if word.lemma in ['']:
                return True
    return False

def place(question_tree):
    root = question_tree.children[0]
    if len(root.children) > 0:
        clause = root.children[0]
        for word in clause.descendants(add_self=True):
            if word.lemma in ['', '', '', '']:
                return True
    return False

def cause(question_tree):
    root = question_tree.children[0]
    if len(root.children) > 0:
        clause = root.children[0]
        for word in clause.descendants(add_self=True):
            if word.lemma in ['', '', '']:
                return True
    return False


def way(question_tree):
    root = question_tree.children[0]
    if len(root.children) > 0:
        clause = root.children[0]
        for word in clause.descendants(add_self=True):
            if word.lemma in ['']:
                return True
    return False

def quantity(question_tree):
    root = question_tree.children[0]
    if len(root.children) > 0:
        clause = root.children[0]
        for word in clause.descendants(add_self=True):
            if word.lemma in ['', '', '']:
                return True
    return False

def whether(question_tree):
    for word in question_tree.descendants(add_self=True):
        if word.lemma in ['']:
            return True
    return False
```
